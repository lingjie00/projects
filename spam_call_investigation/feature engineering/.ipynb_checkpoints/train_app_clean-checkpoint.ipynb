```python
import numpy as np
import pandas as pd
```

```python
train_app_full = pd.read_csv('train/train_app.csv')
```

```python
train_app_full
```

### flow

```python
flow = train_app_full[['phone_no_m','flow']].groupby('phone_no_m').mean().reset_index()
```

```python
flow.columns = ['phone_no_m','network_usage']
```

```python
flow
```

### busi_name

```python
busi_name = train_app_full[['phone_no_m','busi_name']]
```

```python
busi_name['busi_name'].where(busi_name['busi_name'].isna(),'yes',inplace=True)
```

```python
busi_name['busi_name'].where(~busi_name['busi_name'].isna(),'no',inplace=True)
```

```python
busi_name = pd.get_dummies(busi_name,columns=['busi_name'],prefix='data_with_known_app')
```

```python
busi_name = busi_name.groupby('phone_no_m').mean().reset_index()
```

```python
busi_name
```

### train_app_cleaned

```python
train_app = pd.merge(busi_name,flow,on='phone_no_m',how='outer')
```

```python
train_app.to_csv('train/train_app_cleaned.csv',index=False)
```
